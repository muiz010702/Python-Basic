Pengenalan Pyplot

```python
# mengimport modul yang dibutuhkan
import numpy as np
import matplotlib.pyplot as plt
```

Membuat plotting sederhana

```python
plt.plot([3,5,9])
plt.ylabel('sumbu y')
plt.show()
```

```python
plt.plot([8,7,4,2],[5,3,6,5])
plt.ylabel('sumbu y')
plt.xlabel('sumbu x')
plt.show()
```

Pengaturan format pada plot

```python
#mengatur warna dan maker pada plotting
plt.plot([1,3,5,7],[2,4,6,8], 'b*')
plt.axis([0,10,0,10])
plt.show
```

multiple plot dalam satu pemanggilan fungsi plot

```python
M = np.arange(0, 5, 0.5)
M
```

```python
plt.plot(
         M, M, 'r-',
         M, M**2, 'b*',
         M, M**3, 'go')
plt.show()
```

Plotting dengan tipe data kategorical

```python
Rumah = ['Type 3', 'Type 4', 'Type 5']
Harga = [3500, 4500, 6000]
```

```python
plt.plot(Rumah, Harga, 'go')
plt.show()
```

```python
plt.scatter(Harga, Rumah)
plt.show()
```

```python
plt.bar(Rumah, Harga)
plt.show()
```

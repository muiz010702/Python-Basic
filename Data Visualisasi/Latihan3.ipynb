Belajar Multiple Subplots

mengimport library yg dibuthkan

```python
import numpy as np
import matplotlib.pyplot as plt
```

Multiple subplots dengan pyplot style

```python
G1 = np.linspace(0.0, 5.0, 50)
G2 = np.linspace(0.0, 3.0, 50)

M1 = np.cos(2 * np.pi * G1)
M2 = np.cos(2 * np.pi * G2)

G1
```

```python
plt.subplot(2, 1, 1)
plt.plot(G1, G2, 'bo-')
plt.title('Multiple Sublot')
plt.ylabel('Nilai $cos(G_1)$')

plt.subplot(2, 1, 2)
plt.plot(M1, M2, 'g.-')
plt.ylabel('Nilai $cos(G_2)$')
plt.xlabel('Nilai $x$')

plt.show()
```

Mengenal OO Style dalam melakukan plotting

secara explisit membuat figure beserta axes dan memanggil method dari keduanya

```python
x = np.linspace(0, 2, 20)
x
```

```python
fig, ax = plt.subplots() # membuat sebuah figure dan sebuab akses

# melakukan plotting terhadap 3 variane axes
ax.plot(x, x, label= 'Apartment')
ax.plot(x, x**2, label = 'Hotel')
ax.plot(x, x**3, label = 'Vila')

ax.set_xlabel('x label') # menyertakan label x pada axes
ax.set_ylabel('y label') # menyertakan label y pada axes
ax.set_title("Simple Tittle") # menyertakan tittle pada axes
ax.legend() # menyertakan legend pada plotting
```

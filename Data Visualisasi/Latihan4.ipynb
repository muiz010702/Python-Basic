Belajar Grafik Line

```python
# mengimport library yang dibutuhkan
import numpy as np
import matplotlib.pyplot as plt
```

```python
# melakukan persiapan data menggunakan array numpy
A = np.array([8,3,6,2])
A
```

```python
# mendeklarasikan data  kedalam bentuk grafik garis
plt.plot(A, 'b-')
plt.show()
```

```python
# membuat grafik garis dengan array 2 diemnsi
B1 = np.array([5,2,9,8])
B2 = np.array([4,7,3,5])

plt.plot(B1, 'b-', linewidth = '2.0')
plt.plot(B2, 'g-', linewidth = '2.0')
plt.ylabel('Sumbu Y')
plt.xlabel('Sumbu x')
plt.show()
```

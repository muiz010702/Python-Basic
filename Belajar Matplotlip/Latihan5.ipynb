Belajar Label dan title

```python
# mempersiapkan library yang dibutuhkan
import numpy as np
import matplotlib.pyplot as plt
```

```python
# mempersiapkan data data yang dibutuhkan
y1 = np.array([5, 8, 4, 9])
x1 = np.array(['mtk', 'ipa', 'ips', 'indo'])

plt.plot(x1, y1)
plt.title('Hasil nilai')
plt.ylabel('Nilai')
plt.xlabel('Mata Kuliah')
plt.show()
```

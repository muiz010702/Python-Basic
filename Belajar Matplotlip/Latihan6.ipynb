Belajar Scatter Plot

```python
# Mempersiapkan library yang dibutuhkan

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Mempesiapkan dataset yang dibutuhkan menggunakan array

A = np.array([20,30,40,50,60])
B = np.array([10,9,8,7,6])
```

```python
# Membuat grafik scatter sederhana menggunakan fungsi scatter()
plt.scatter(A, B)
plt.show()
```

```python
C = np.array([50,35,25,48,56,70,32])
D = np.array([10,3,15,45,23,65,45])
plt.scatter(C, D, color = 'green', alpha=0.25)

E = np.array([15,24,43,25,60])
F = np.array([43,25,12,18,76])
plt.scatter(E, F, color = 'yellow', alpha=0.25)

plt.show()
```

Renaming (Mengubah)

Fungsi pertama yang akan kami perkenalkan di sini adalah rename(), yang memungkinkan Anda mengubah nama indeks dan/atau nama kolom. Misalnya, untuk mengubah kolom Total di kumpulan data menjadi Jumlah, kita akan melakukan:

```python
# contoh 1

dataset.rename(columns={'Total':'Jumlah'})
# sintaks ini digunakan untuk mengubah kolom Total di kumpulan data menjadi Jumlah
```

rename() memungkinkan Anda mengganti nama nilai indeks atau kolom dengan menentukan masing-masing parameter kata kunci indeks atau kolom. Ini mendukung berbagai format masukan, tetapi biasanya kamus Python adalah yang paling nyaman. Berikut adalah contoh penggunaannya untuk mengganti nama beberapa elemen indeks.

```python
# contoh 2

dataset.rename(index={0:'Satu', 1:'Dua', 2:'Tiga'})
# sintaks ini digunakan untuk mengganti suatu indeks tertentu ke dalam string
```

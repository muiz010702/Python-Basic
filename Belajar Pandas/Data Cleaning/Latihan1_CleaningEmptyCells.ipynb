Pembersihan sel kossong

Sel kosong berpotensi memberikan hasil yang salah saat Anda menganalisis data.

1, Hapus Baris

#Salah satu cara mengatasi sel kosong adalah dengan menghilangkan baris yang berisi sel kosong.

#Hal ini biasanya tidak masalah, karena kumpulan data bisa berukuran sangat besar
dan menghapus beberapa baris tidak akan berdampak besar pada hasilnya.

```python
# mempersiapkan library yang dibutuhkan
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# contoh 1
# Mengembalikan Bingkai Data baru tanpa sel kosong:

Data = pd.read_csv('C:\Latihan\Latihan\Python\Belajar Pandas\Data Cleaning\Sample Data\data.csv')
NewData = Data.dropna()

print(NewData.to_string())

# Catatan: Secara default, metode dropna() mengembalikan DataFrame baru dan
# tidak akan mengubah yang asli.
```

```python
# Jika Anda ingin mengubah DataFrame asli, gunakan argumen inplace = True :
# Hapus semua baris dengan nilai NULL:

Data = pd.read_csv('C:\Latihan\Latihan\Python\Belajar Pandas\Data Cleaning\Sample Data\data.csv')
Data.dropna(inplace=True)

print(Data.to_string())
```

```python
#Perhatikan pada hasil bahwa beberapa baris telah dihapus (baris 18, 22 dan 28).
#Baris ini memiliki sel dengan nilai kosong.

# Catatan: Sekarang, dropna(inplace = True) TIDAK akan mengembalikan DataFrame baru, tetapi akan menghapus 
# semua baris yang berisi nilai NULL dari DataFrame asli.
```

2. Ganti Nilai Kosong

Cara lain untuk menangani sel kosong adalah dengan memasukkan nilai baru.
Dengan cara ini Anda tidak perlu menghapus seluruh baris hanya karena beberapa sel kosong.
Metode fillna() memungkinkan kita mengganti sel kosong dengan nilai:

```python
# contoh 2

# Ganti nilai NULL dengan angka 130:
Data = pd.read_csv('C:\Latihan\Latihan\Python\Belajar Pandas\Data Cleaning\Sample Data\data.csv')
Data.fillna(130, inplace=True)

print(Data.to_string())
```

```python
#Perhatikan hasilnya: sel kosong mendapat nilai 130 (di baris 18, 22 dan 28).
```

```python
# jika anda hanya ingin mengganti nilai pada kolom tertentu
# Contoh di atas menggantikan semua sel kosong di seluruh Data Frame.
# Untuk hanya mengganti nilai kosong pada satu kolom, tentukan nama kolom untuk DataFrame:
```

```python
# contoh 3

# Ganti nilai NULL pada kolom "Kalori" dengan angka 130:
Data = pd.read_csv('C:\Latihan\Latihan\Python\Belajar Pandas\Data Cleaning\Sample Data\data.csv')
Data['Calories'].fillna(130, inplace=True)

print(Data.to_string())
```

```python
#Operasi ini memasukkan 130 ke dalam sel kosong di kolom "Kalori" (baris 18 dan 28).
```

3. Ganti Menggunakan Mean, Median, atau Mode
Cara umum untuk mengganti sel kosong adalah dengan menghitung nilai mean, median, atau mode kolom.
Pandas menggunakan metode mean() median() dan mode() untuk menghitung nilai masing-masing untuk kolom tertentu:

```python
# contoh menggunakan metode mean()
# Hitung MEAN, dan ganti nilai kosong dengan nilai tersebut:
# Mean = nilai rata-rata (jumlah seluruh nilai dibagi banyaknya nilai).

Data = pd.read_csv('C:\Latihan\Latihan\Python\Belajar Pandas\Data Cleaning\Sample Data\data.csv')
X = Data["Calories"].mean()
Data["Calories"].fillna(X, inplace=True)

print(Data.to_string())
```

```python
# contoh menggunakan metode median()
# Hitung median, dan ganti nilai kosong dengan nilai tersebut:
# Median = nilai yang berada di tengah, setelah Anda mengurutkan semua nilai secara menaik.

Data = pd.read_csv('C:\Latihan\Latihan\Python\Belajar Pandas\Data Cleaning\Sample Data\data.csv')
X = Data["Calories"].median()
Data["Calories"].fillna(X, inplace=True)

print(Data.to_string())
```

```python
# contoh menggunakan metode mode()
# Hitung MODE, dan ganti nilai kosong dengan itu:
# Mode = nilai yang paling sering muncul.

Data = pd.read_csv('C:\Latihan\Latihan\Python\Belajar Pandas\Data Cleaning\Sample Data\data.csv')
X = Data["Calories"].mode()[0]
Data["Calories"].fillna(X, inplace=True)

print(Data.to_string())
```
